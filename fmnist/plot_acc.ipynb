```python
# preamble
import os
import sys
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import time

print('Now: ' + time.strftime('%c'))
print(sys.version)

# load data

m_10_ps = np.loadtxt('./logs/acc_c-0.txt', comments='#', delimiter=' ', usecols = (1))
m_50_ps = np.loadtxt('./logs/acc_c-1.txt', comments='#', delimiter=' ', usecols = (1))
m_100_ps = np.loadtxt('./logs/acc_c-2.txt', comments='#', delimiter=' ', usecols = (1))
m_500_ps = np.loadtxt('./logs/acc_c-3.txt', comments='#', delimiter=' ', usecols = (1))

m_10_di = np.loadtxt('./logs/acc_c-10.txt', comments='#', delimiter=' ', usecols = (1))
m_50_di = np.loadtxt('./logs/acc_c-11.txt', comments='#', delimiter=' ', usecols = (1))
m_100_di = np.loadtxt('./logs/acc_c-12.txt', comments='#', delimiter=' ', usecols = (1))
m_500_di = np.loadtxt('./logs/acc_c-13.txt', comments='#', delimiter=' ', usecols = (1))

m_10_sym = np.loadtxt('./logs/acc_c-20.txt', comments='#', delimiter=' ', usecols = (1))
m_50_sym = np.loadtxt('./logs/acc_c-21.txt', comments='#', delimiter=' ', usecols = (1))
m_100_sym = np.loadtxt('./logs/acc_c-22.txt', comments='#', delimiter=' ', usecols = (1))
m_500_sym = np.loadtxt('./logs/acc_c-23.txt', comments='#', delimiter=' ', usecols = (1))

m_10_dst = np.loadtxt('./logs/acc_c-30.txt', comments='#', delimiter=' ', usecols = (1))
m_50_dst = np.loadtxt('./logs/acc_c-31.txt', comments='#', delimiter=' ', usecols = (1))
m_100_dst = np.loadtxt('./logs/acc_c-32.txt', comments='#', delimiter=' ', usecols = (1))
m_500_dst = np.loadtxt('./logs/acc_c-33.txt', comments='#', delimiter=' ', usecols = (1))
```

```python
# common plot parameters
plt.rcParams.update({'font.size': 16})
plt.rcParams["figure.figsize"] = (15,10)
fig, ax = plt.subplots(2, 2)

x_axis = ['100', '500', '1000', '5000']
texts = ['Weakest model', 'Second model', 'Third model', 'Strongest model']
base_d = np.loadtxt('./logs/acc_c-40.txt', comments='#', delimiter=' ', usecols = (1))

for i in range(4):
    m_ps = [m_10_ps[i], m_50_ps[i], m_100_ps[i], m_500_ps[i]]
    m_di = [m_10_di[i], m_50_di[i], m_100_di[i], m_500_di[i]]
    m_sym = [m_10_sym[i], m_50_sym[i], m_100_sym[i], m_500_sym[i]]
    m_dst = [m_10_dst[i], m_50_dst[i], m_100_dst[i], m_500_dst[i]]

    ax[i//2, i % 2].plot(x_axis, m_ps);
    ax[i//2, i % 2].plot(x_axis, m_di);
    ax[i//2, i % 2].plot(x_axis, m_sym);
    ax[i//2, i % 2].plot(x_axis, m_dst);
    ax[i//2, i % 2].plot(x_axis, [base_d[i]] * 4);

    ax[i//2, i % 2].grid(True);
    ax[i//2, i % 2].set_ylim((0.5, 0.92));
    ax[i//2, i % 2].title.set_text(texts[i])

plt.legend(['discriminative', 'distillation', 'symmetric', 'cooperative', 'optimal']);
```
